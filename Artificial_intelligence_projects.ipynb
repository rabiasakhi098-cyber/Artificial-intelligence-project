```python
# AI Chatbot for College Helpdesk

def chatbot():
    print("🎓 Welcome to College Helpdesk Chatbot!")
    print("Type 'exit' to end chat.\n")

    while True:
        user = input("You: ").lower()

        if user == "exit":
            print("Bot: Thank you! Have a great day.")
            break

        # Admission related
        elif "admission" in user or "apply" in user:
            print("Bot: Admissions are open. You can apply online through the college website.")

        # Fees related
        elif "fee" in user or "fees" in user:
            print("Bot: Please visit the accounts section or website for fee details.")

        # Exam related
        elif "exam" in user:
            print("Bot: Exams start next month. Check the notice board for schedule.")

        # Result related
        elif "result" in user:
            print("Bot: Results will be announced on the student portal.")

        # Campus facilities
        elif "library" in user:
            print("Bot: Library is open from 9 AM to 5 PM.")

        elif "hostel" in user:
            print("Bot: Hostel facility is available for boys and girls.")

        elif "canteen" in user:
            print("Bot: Canteen is open from 8 AM to 6 PM.")

        # Greetings
        elif "hello" in user or "hi" in user:
            print("Bot: Hello! How can I help you today?")

        # Default reply
        else:
            print("Bot: Sorry, I didn't understand. Please ask about admissions, exams, fees, hostel, library etc.")

# Run chatbot
chatbot()
```

```python
# Tic-Tac-Toe with  (Minimax)AI
import math

# Create board
board = [" " for _ in range(9)]

def print_board():
    for i in range(0, 9, 3):
        print(board[i] + " | " + board[i+1] + " | " + board[i+2])
        if i < 6:
            print("--+---+--")

def is_winner(b, player):
    win_conditions = [
        [0,1,2], [3,4,5], [6,7,8],  # rows
        [0,3,6], [1,4,7], [2,5,8],  # columns
        [0,4,8], [2,4,6]            # diagonals
    ]
    return any(all(b[i] == player for i in cond) for cond in win_conditions)

def is_full(b):
    return " " not in b

def minimax(b, depth, is_maximizing):
    if is_winner(b, "O"):   # AI
        return 1
    if is_winner(b, "X"):   # Human
        return -1
    if is_full(b):
        return 0

    if is_maximizing:
        best = -math.inf
        for i in range(9):
            if b[i] == " ":
                b[i] = "O"
                score = minimax(b, depth+1, False)
                b[i] = " "
                best = max(best, score)
        return best
    else:
        best = math.inf
        for i in range(9):
            if b[i] == " ":
                b[i] = "X"
                score = minimax(b, depth+1, True)
                b[i] = " "
                best = min(best, score)
        return best

def ai_move():
    best_score = -math.inf
    move = -1
    for i in range(9):
        if board[i] == " ":
            board[i] = "O"
            score = minimax(board, 0, False)
            board[i] = " "
            if score > best_score:
                best_score = score
                move = i
    board[move] = "O"

def play_game():
    print("Positions are 0 to 8 (left to right)")
    print_board()

    while True:
        # Human move
        pos = int(input("Enter your move (0-8): "))
        if board[pos] != " ":
            print("Invalid move! Try again.")
            continue

        board[pos] = "X"
        print_board()

        if is_winner(board, "X"):
            print("You win!")
            break
        if is_full(board):
            print("It's a draw!")
            break

        # AI move
        ai_move()
        print_board()

        if is_winner(board, "O"):
            print("AI wins!")
            break
        if is_full(board):
            print("It's a draw!")
            break

# Run the game
play_game()
```

```python
# Easy Face Recognition Attendance System
# Install first:
# pip install opencv-python pandas

import cv2
import pandas as pd
from datetime import datetime

# Load face detector
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +
'haarcascade_frontalface_default.xml')

# Create attendance file
file_name = "Attendance.csv"

# Start webcam
cap = cv2.VideoCapture(0)

print("Press S = Mark Attendance")
print("Press Q = Quit")

while True:
    ret, frame = cap.read()

    # Check if frame was read successfully
    if not ret:
        print("Error: Failed to capture frame from webcam. Exiting.")
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    # Draw rectangle around face
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y),
                      (x+w, y+h),
                      (0, 255, 0), 2)

    cv2.imshow("Attendance System", frame)

    key = cv2.waitKey(1)

    # Press S to save attendance
    if key == ord('s'):
        name = input("Enter Student Name: ")
        time_now = datetime.now().strftime("%H:%M:%S")
        date_now = datetime.now().strftime("%d-%m-%Y")

        data = {
            "Name": [name],
            "Date": [date_now],
            "Time": [time_now]
        }

        df = pd.DataFrame(data)

        try:
            old = pd.read_csv(file_name)
            df = pd.concat([old, df], ignore_index=True)
        except:
            pass

        df.to_csv(file_name, index=False)
        print("Attendance Marked!")

    # Press Q to quit
    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
